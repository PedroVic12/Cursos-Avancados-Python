```python
x = [1, 2, 3, 4]
y = [3, 4, 5, 6]
plt.plot(x, y, label = "dados")


#EIXOS
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

#Mudan√ßa de escala
plt.title("Titulo do grafico")
plt.xticks([0,2,4,5,8])
plt.yticks([-1,3,5,9,11])

#Valor da legenda
plt.legend()

plt.show()
```

```python
x1 = np.linspace(0,5,10)
y1 = x1**2
plt.plot(x1,y1)

#Titulo:
plt.title("Titulo do gr√°fico")

#Eixos:
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

```

### **V√°rios gr√°ficos juntos** (Subplots)

```python
valores_x = [1.2,4,5]
valores_y = [1.4,3,2]

# Grafico 1:
figura = plt.figure(figsize=(20,8))
figura.suptitle("Titulo geral") 
figura.add_subplot(131)
plt.plot(valores_x, valores_y, label='Um dado qualquer')
plt.legend()
plt.title("Grafico 1")

#Grafico 2:
figura.add_subplot(132)
plt.bar(valores_x, valores_y)
plt.title("Grafico 2")

#Grafico 3:    
figura.add_subplot(133)  # (1 linha, 3 coluna, 3 posi√ß√£o)
plt.scatter(valores_x, valores_y)
plt.title('Grafico 3')


plt.show(figura)
```

### **Usando Figuras como objetos**

```python
#Criando figuras
fig_1 = plt.figure(figsize=(5,4),dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])

#Primeiro gr√°fico
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title('Titulo')
axes_1.plot(x1,y1,label='x/x2')
axes_1.plot(y1,x1,label='x2/x')
axes_1.legend(loc=0)


#Segundo Gr√°fico
axes_2 = fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days squared')
axes_2.set_title('Titulo')
axes_2.plot(x1,y1,label='x/x2')

axes_2.text(0,40,"TEXTO TESTE",color='red')
```

- Buscando so a bolsa de nova york (NYA)

```python
exemplo_df2 = exemplo_df[['Index', 'Date','Open']]
display(exemplo_df2)
```

```python
dfremove = exemplo_df2.loc[(exemplo_df['Index'] != 'NYA')]

cotacao_novaYork_df = exemplo_df2.drop(dfremove.index)

cotacao_novaYork_df = cotacao_novaYork_df[13900:]

display(cotacao_novaYork_df)
```

```python
plt.figure(figsize=(20,4))
plt.plot(cotacao_novaYork_df['Date'], cotacao_novaYork_df['Open'],label = 'Cota√ß√µes do d√≥lar')
plt.legend(loc= 0, fontsize=18)
plt.ylabel('Valor do fechamento')
plt.xlabel('Data da cota√ß√£o')
plt.title('Grafico de cota√ß√£o da Bolsa de Nova York')
plt.xticks(['2021-03-23', '2021-04-23', '2021-05-23'])

plt.show()
```
